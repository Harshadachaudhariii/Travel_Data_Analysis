```python
# data is either numerical or categorical, numerical can be either discrete and continuous 
# discrete data through its numerical but should be treated as categorical to do the analysis
# but for model building all the categorical data should be converted to numbers
```

```python
num_feature = X.select_dtypes(exclude='object').columns   # ✅ numeric
cat_feature = X.select_dtypes(include='object').columns   # ✅ categorical
```

```python
# scaling is done for numerical data 
# categorical is done by onehotendcoing
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer 
```

```python
numeric_transform = StandardScaler()
on_encoding_transform = OneHotEncoder(drop='first')
preprocessor = ColumnTransformer(
    [
        ('OneHotEncoder',on_encoding_transform, cat_feature ),
        ('StandardScaler',numeric_transform, num_feature )
    ]
)
```

```python
preprocessor
```

```python
X_train = preprocessor.fit_transform(X_train)
pd.DataFrame(X_train)
```

```python
X_test = preprocessor.transform(X_test)
pd.DataFrame(X_test)
```
